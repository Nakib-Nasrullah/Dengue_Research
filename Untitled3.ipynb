```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# 1. Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Datasets/Dengue_final_dataset_updated.csv')
```

```python
# 2. Data Preprocessing
# Convert date and sort to maintain temporal order
df['S'] = pd.to_datetime(df['S'])
df = df.sort_values(by=['District', 'S'])

# Encode categorical variables (District and Division)
le_district = LabelEncoder()
df['District_encoded'] = le_district.fit_transform(df['District'])
le_division = LabelEncoder()
df['Division_encoded'] = le_division.fit_transform(df['Division'])
```

```python
# 3. Feature Selection
# Including weather parameters, lagged features, and demographics
features = [
    'Max_Temp', 'Min_Temp', 'Avg_Temp', 'Humidity', 'Rainfall',
    'Rainfall_Lag_7', 'Rainfall_Lag_14', 'Rainfall_Lag_21', 'Rainfall_Lag_30',
    'Temp_Lag_7', 'Humidity_Lag_7', 'Humidity_Lag_14', 'Humidity_Lag_21',
    'Humidity_Lag_30', 'Dengue_Lag_7', 'Population_Density', 'Urban_Index',
    'District_encoded', 'Division_encoded'
]
target = 'Dengue_case'
```

```python
# 4. Scaling
# Scaling features and target separately to allow for inverse transformation later
scaler_features = MinMaxScaler()
scaler_target = MinMaxScaler()

df_scaled_features = scaler_features.fit_transform(df[features])
df_scaled_target = scaler_target.fit_transform(df[[target]])
```

```python
# 5. Create Time-Series Sequences (Windowing)
def create_sequences(data, target, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:(i + window_size)])
        y.append(target[i + window_size])
    return np.array(X), np.array(y)

# We use a 14-day look-back period (window_size = 14)
window_size = 14
X, y = create_sequences(df_scaled_features, df_scaled_target, window_size)
```

```python
# 6. Train-Test Split
# Temporal split: using the first 80% for training and last 20% for testing
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# 7. Build the LSTM Model
model = Sequential([
    # First LSTM layer with Dropout for regularization
    LSTM(64, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    # Second LSTM layer
    LSTM(32, activation='relu'),
    Dropout(0.2),
    # Final dense layer for case count prediction
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

```python
# 8. Train the Model
# Validation split helps monitor overfitting during training
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```
