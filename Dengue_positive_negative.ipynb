```python
# Convert and sort date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
# Calculate daily change
df['Daily_Change'] = df['Dengue_case'].diff()
```

```python
# Assign trend labels
df['Trend'] = df['Daily_Change'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'No Change')
)

print(df[['Date','Dengue_case','Daily_Change','Trend']].head())
```
