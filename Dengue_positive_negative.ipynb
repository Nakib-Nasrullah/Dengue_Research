# **Data Preparation**

```python
# Preprocess
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
# Create trend feature
df['Daily_Change'] = df['Dengue_case'].diff().fillna(0)
df['Trend_Label'] = df['Daily_Change'].apply(
    lambda x: 1 if x > 0 else (0 if x == 0 else -1)
)
```

```python
# Fill missing values if any
df = df.fillna(method='ffill')
```

# **Visualization**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df['Date'], df['Dengue_case'])
plt.title("Dengue Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.show()
```

# **Heatmap (Feature Correlation)**

```python
import seaborn as sns

features = ['Avg_Temp', 'Humidity', 'Rainfall', 'Trend_Label', 'Dengue_case']

corr = df[features].corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```
